```python
import pandas as pd
from pandas_datareader import data, wb
import datetime
import yfinance as yf
yf.pdr_override()

# Bank of America
BBCA = data.get_data_yahoo('BBCA.JK')
BBCA.tail()
```

```python
df = pd.DataFrame(BBCA)
df.reset_index(inplace = True)
df.tail()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Convert to datetime format
df['Date'] = pd.to_datetime(df['Date'])

#Filter from 2022 until now
df_filtered = df[df['Date'] >= '2022-01-01']

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_filtered, x='Date', y='Close')
plt.title('BBCA Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Convert to datetime format
df['Date'] = pd.to_datetime(df['Date'])

#Filter from 2022 until now
df_filtered = df[df['Date'] >= '2022-01-01']

#Further filter the DataFrame for the specific date range for the yellow line
df_highlight = df_filtered[(df_filtered['Date'] >= '2024-05-01') & (df_filtered['Date'] <= '2024-05-29')]

# Set up the figure and axes
plt.figure(figsize=(10, 6))

# Create the line chart for the entire dataset
sns.lineplot(data=df_filtered, x='Date', y='Close', label='BBCA Stock Price', color='#1F77B4')

# Overlay the line chart for the highlighted date range
sns.lineplot(data=df_highlight, x='Date', y='Close', label='Highlighted Period', color='#FFF200')

# Set the title and labels
plt.title('BBCA Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')

# Rotate x-axis labels for better readability
plt.xticks(rotation=0)

# Show the legend
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
